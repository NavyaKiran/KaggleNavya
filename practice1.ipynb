```python
#Load the data
import pandas as pd 

file_path = '/kaggle/input/melbourne-housing-market/Melbourne_housing_FULL.csv'
home_data = pd.read_csv(file_path)
home_data.head(5)
```

```python
#Review the summary stats 
home_data.describe()
#Count is the number of non zero values in that column 
#mean is the average of values present 
#std is the standard deviation or the amount by which the values are spread out 
#min is the minimum value in that column 
#25% is the value present at the 25th percentile and same for 50 and 75 
#max is the maximum value in that column
```

```python
home_data.columns
#list out all the colums available in the dataset
```

```python
home_data = home_data.dropna(axis = 0)
home_data.head()
```

```python
home_data.isna().sum()
home_data.isnull().sum()
```

```python
y = home_data.Price
y.head()
```

```python
features = ['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude']
X = home_data[features]
```

```python
from sklearn.tree import DecisionTreeRegressor

model = DecisionTreeRegressor(random_state=1)
model.fit(X, y)
```

```python
#Make predictions for 10 houses 
X_pred = X[:10]
print(X_pred)
```

```python
print(model.predict(X_pred))
```
